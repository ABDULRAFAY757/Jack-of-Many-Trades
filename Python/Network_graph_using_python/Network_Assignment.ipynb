```python
import matplotlib.pyplot as plt
import networkx as nx
j=nx.Graph()
j.add_node('Arad',pos=(-4,-20))
j.add_node('Zerind',pos=(1,13))
j.add_node('Timisoara',pos=(-2,-70))
j.add_node('Sibiu',pos=(4,-40))
j.add_node('Oradea',pos=(6,50))
j.add_node('Lugoj',pos=(2,-100))
j.add_node('Mehadia',pos=(2,-150))
j.add_node('Drobeta',pos=(2,-200))
j.add_node('Fagras',pos=(12,-40))
j.add_node('Rimnicu',pos=(9,-100))
j.add_node('Pitesti',pos=(10,-200))
j.add_node('Craiova',pos=(7,-230))
j.add_node('Bucharest',pos=(15,-150))
j.add_node('Giurgiu',pos=(20,-200))
j.add_node('Urziceni',pos=(20,-90))
j.add_node('Hirsova',pos=(28,-130))
j.add_node('Eforie',pos=(28,-170))
j.add_node('Vaslui',pos=(20,-50))
j.add_node('Iasi',pos=(30,-50))
j.add_node('Neamt',pos=(40,-50))

j.add_edge('Arad','Zerind',weight=75)
j.add_edge('Arad','Timisoara',weight=118)
j.add_edge('Arad','Sibiu',weight=140)
j.add_edge('Zerind','Oradea',weight=71)
j.add_edge('Oradea','Sibiu',weight=151)
j.add_edge('Timisoara','Lugoj',weight=111)
j.add_edge('Lugoj','Mehadia',weight=70)
j.add_edge('Mehadia','Drobeta',weight=75)
j.add_edge('Drobeta','Craiova',weight=120)
j.add_edge('Craiova','Pitesti',weight=138)
j.add_edge('Craiova','Rimnicu',weight=146)
j.add_edge('Rimnicu','Sibiu',weight=80)
j.add_edge('Sibiu','Fagras',weight=211)
j.add_edge('Fagras','Bucharest',weight=99)
j.add_edge('Pitesti','Bucharest',weight=101)
j.add_edge('Bucharest','Giurgiu',weight=90)
j.add_edge('Bucharest','Urziceni',weight=85)
j.add_edge('Urziceni','Hirsova',weight=98)
j.add_edge('Hirsova','Eforie',weight=86)
j.add_edge('Urziceni','Vaslui',weight=142)
j.add_edge('Vaslui','Iasi',weight=92)
j.add_edge('Iasi','Neamt',weight=87)
pos=nx.get_node_attributes(j,'pos')
nx.draw(j,pos,with_labels=True,node_size=1000,node_color="pink",alpha=1,edge_color="Black")

#pos=nx.spring_layout(j)
labels=nx.get_edge_attributes(j,'weight')
nx.draw_networkx_edge_labels(j,pos,edge_labels=labels)
bb = nx.edge_betweenness_centrality(j, normalized=True)
nx.set_edge_attributes(j, bb, "betweenness")
#fg=plt.figure(5,figsize=(5,5))

plt.rcParams['figure.figsize']=[6,8]
plt.rcParams["figure.autolayout"]=True
plt.savefig("Romaina.png")
plt.title("Romania_Map")

```
